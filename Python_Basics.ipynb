# Python Basics Summary

## 1-Variables

* Variables are containers that store the data
* Python variables datatypes are (int - float - str - bool)
* Python variables datatypes are dynamic (which means that you don't need to define it).
* Adding two variables by using (+) and Substracting by (-)
* Multipling two variables by (*) and deviding by (/)

```python
#Declaring a variable and printing its value
num1=10
print(num1)
print("-"*20)
#Adding two numbers and printing the result
num2=20
total = num1 + num2
print(f"The sum of {num1} and {num2} is {total}")
```

## 2-Data Structures 
There are 4 built-in data structures in python and the i will take a breif for each one.
* Lists : [-,-,-,-]
  * Indexed
  * Mutable
  * Ordered
  * Contain Duplicates
* Tuples : (-,-,-,-)
  * Indexed
  * Immutable
  * Ordered
  * Contain Duplicates
* Sets : {-,-,-,-}
  * Non-indexed
  * Immutable
  * Unordered
  * Unique Values
* Dictionaries : {k1:v1 , k2:v2 , k3:v3 , k4:v4}
  * Key-Value paired
  * Mutable
  * Ordered by keys
  * Unique Keys

```python
#Declaring a list and adding elements to it
my_list = []
my_list.append(1)
my_list.append(2)
my_list.append('Ahmed Shahin')
print(my_list)
name=my_list[2]
print(name)
```

```python
# When adding two lists, looks like append each element of the second list to the first list.
list1 = [1, 2, 3, 4, 5]
list2 = [1, 2, 3, 4, 5]
list3 = list1 + list2
print(list3)
```

```python
#Declaring a Tuple with multiple values and printing a specific value from it
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple[2])
```

```python
mylist = [1, 2, 3, 4, 5,2,3]
print("List:", mylist)
print("Length of the list:", len(mylist),end="\n\n\n")
# Converting a list to a set to remove duplicates
myset = set(mylist)
print("Set:", myset)
print("Length of the set:", len(myset))
```

```python
# Declaring a dictionary and printing its values
my_dict = {"name": "Ahmed Shahin", "age": 25, "city": "Kafr El Sheikh"}
print("Dictionary:", my_dict)
print("Name:", my_dict["name"])
print("Age:", my_dict["age"])
print("Keys:", list(my_dict.keys()))
print("Values:", list(my_dict.values()))
print("Items:", list(my_dict.items()))
```

# 4-Control Flow in Python

* Conditionals
  * If - Elif - Else 
* Loops
  * For
  * While


```python
# Conditional statements for checking if a number is positive, negative or zero
num = int(input("Enter a number: "))
if num > 0:
    print(f"{num} is a positive number.")
elif num == 0:
    print(f"{num} is zero.")
else:
    print(f"{num} is a negative number.")
```

```python
#Printing all elements of a list using a for loop and condational statement
mylist = [1,65,6,'hello',1952]
for i in mylist:
    if (type(i) is int) and (i > 1000):
        continue
    print(i)
```

```python
#
num = 15
while num > 0:
    if num == 4:
        break
    print(num)
    num -= 1
```
